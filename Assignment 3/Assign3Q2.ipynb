```python
with open("/content/drive/MyDrive/tokens_of_hindi_corpus.txt") as f:
    words = [line.strip() for line in f]

frequency_map = {}

for word in words:
    if word in frequency_map:
       frequency_map[word] += 1
    else:
        frequency_map[word] = 1
```

```python
items = list(frequency_map.items())
items.sort(key=lambda x: x[1], reverse=True)
num_to_print = 100
lis = []

print(f"First {num_to_print} entries of the frequency map:")
for i in range(num_to_print):
    if i < len(items):
        lis.append({items[i][1],items[i][0]})
        print(f"'{items[i][0]}': {items[i][1]}")
    else:
        break
```

```python
import matplotlib.pyplot as plt
from matplotlib import rcParams

rcParams['font.family'] = 'Nirmala UI'

pairs = []
for s in lis:
    word = None
    freq = None
    for item in s:
        if isinstance(item, int):
            freq = item
        else:
            word = item
    pairs.append((word, freq))

words, freqs = zip(*pairs)

# Plot
plt.figure(figsize=(40, 8))
plt.bar(words, freqs, color='skyblue')

plt.xlabel("Word", fontsize=14)
plt.ylabel("Frequency", fontsize=14)
plt.title("Top Frequent Words", fontsize=16)

plt.xticks(rotation=0, fontsize=17)

plt.tight_layout()
plt.show()
```

```python

pairs = []
for s in lis:
    word = None
    freq = None
    for item in s:
        if isinstance(item, int):
            freq = item
        else:
            word = item
    pairs.append((word, freq))

words, freqs = zip(*pairs)

# Define thresholds
thresholds = [65000, 50000, 30000]

for t in thresholds:
    # Filter based on threshold
    filtered = [(w, f) for w, f in zip(words, freqs) if f < t]

    if not filtered:
        print(f"No words with freq < {t}")
        continue

    fw, ff = zip(*filtered)



    plt.figure(figsize=(40, 8))
    plt.bar(fw, ff, color='skyblue')

    plt.xlabel("Word", fontsize=14)
    plt.ylabel("Frequency", fontsize=14)

    plt.title(f"Words with Frequency < {t}", fontsize=14)
    plt.xticks(rotation=0, fontsize=17)

    plt.tight_layout()
    plt.show()
```
